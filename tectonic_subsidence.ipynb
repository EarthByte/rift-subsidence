```python
%matplotlib inline
import matplotlib.pyplot as plt

from tectonic_subsidence import TectonicSubsidence

def my_plot(ax,ts):
    ts.plot_sea_level(ax)
    ts.plot_tectonic_subsidence(ax)
    ts.plot_dynamic_topography(ax)
    ts.plot_subsidence_minus_dynamic_topography(ax)
    ts.plot_rift_subsidence(ax,0)
    ts.plot_residual(ax,0)
    ts.config_ax(ax)
    ax.set_title(f'Tectonic Subsidence')

fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(nrows=2, ncols=2,figsize=(16,12))
ts1 = TectonicSubsidence('sunrise', 'Haq87_SealevelCurve_Longterm', 'M6', 'parameters.json', 'my_results')
ts2 = TectonicSubsidence('sunrise', 'Haq87_SealevelCurve_Longterm', 'M7', 'parameters.json', 'my_results')
ts3 = TectonicSubsidence('swift', 'Haq87_SealevelCurve_Longterm', 'M6', 'parameters.json', 'my_results')
ts4 = TectonicSubsidence('sunrise', 'Haq87_SealevelCurve_Longterm', 'M6', 'parameters.json', 'my_results', 100)

my_plot(ax1,ts1)
my_plot(ax2,ts2)
my_plot(ax3,ts3)
my_plot(ax4,ts4)

plt.show()

fig.savefig(f'my_figure.png')
```
